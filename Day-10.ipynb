```python
import pandas as pd
import seaborn as sns
```

```python
dataset = pd.read_excel("general_data.xlsx",sheet_name = 0)
```

# Hypothesis 1

```python
sns.countplot(x = "Age",hue = "Attrition",data = dataset)
```

From the above graph we can conclude that:
--> Employee having age greater than or equal to 36 are having low attrition rate
-->Ho : µ >= 36(Null Hypothesis)
   H1 : µ < 36 (Alternative Hypothesis)
